```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Dataset_of_Diabetes.csv')
```

```python
data.shape
```

```python
data.info()
```

```python
data.columns
```

```python
###DROPPING COLUMN
data.drop(columns = ['TG', 'No_Pation'], axis=1, inplace=True)
```

```python
data.columns
```

```python
###CHANGING COLUMN NAMES
data = data.rename(columns = {'AGE': 'Age_in_years', 'Cr': 'Creatinine', 'Chol': 'Cholesterol', 'CLASS': 'Class'})
data
```

```python
##DROPPING DUPLICATES
data.drop_duplicates()
```

```python
##REMOVING NULL
data.dropna()
```

```python
#OUTLIERS-1
sns.boxplot(data['BMI'])
```

```python
upper_limit = data['BMI'].mean() + 3*data['BMI'].std()
lower_limit = data['BMI'].mean() - 3*data['BMI'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
##FINDING OUTLIERS
data.loc[(data['BMI'] > upper_limit) | (data['BMI'] < lower_limit)]
```

```python
##DELETING OUTLIERS
new_data1 = data.loc[(data['BMI'] < upper_limit) & (data['BMI'] > lower_limit)]
print('before removing outliers:', len(data))
print('after removing outliers:', len(new_data1))
print('outliers:', len(data)-len(new_data1))
```

```python
sns.boxplot(new_data1['BMI'])
```
